```python
for i in websites:
    response = requests.get(i)
    soup = BeautifulSoup(response.content,'html.parser')
    results = soup.find(id = 'content') 
    s = results.find('h1')
    print(s.text)
    tds = []
    b = []
    c = []
    d = []
    e = []
    fourth = []
    g = []
    y = []
    x = []
    divs = soup.findAll('table', {'class':'properties'})
    for div in divs:
        rows = div.findAll('tr')
        for row in rows:
            tds.append(row.text)
    print(tds)
    for i in tds:
        w = i.find('curve fit')
        if tds.index(i)<16:
            if w!=-1:
                r = tds.index(i)
                x.append(r)
                #print(i)
                #print(x)
            else:
                pass
    #print(tds[13])
    print(x)
    for i in range(1,14):
            #print(i)
            a = str.split(tds[i], sep='\n')
            b.append(a[1])
            c.append(a[2])
            d.append(a[3])
            if len(a)>4:
                e.append(a[4])
            if len(a)>5:
                fourth.append(a[5])
    #print(a)
    #print(b)
    #print(c)
    #print(d)
    #print(e)
    #print(fourth)
    
    tuples = [(key, value) for i, (key, value) in enumerate(zip(b, c))]
    res = dict(tuples)
    #print(res)
    
    key = list(res.keys())
    val = list(res.values())
    #print(key)
    t = key.index('data range')
    #print(t)
    keys1 = []
    vals1 = []
    for i in range(0,t):
            keys1.append(key[i])
            vals1.append(val[i])
    print(keys1)
    print(vals1)
    keys2 = []
    vals2 = []
    for i in range(t,t+2):
            keys2.append(key[i])
            f = str.split(val[i], sep='-')
            h = (f[0],f[1])
            vals2.append(h)
    print(keys2)
    print(vals2)
    keys3 = []
    vals3 = []
    for i in range(t+2,len(res)):
            keys3.append(key[i])
            vals3.append(val[i])
    print(keys3)
    print(vals3)
    print()

    #the equation in the website
    formula = divs = soup.findAll('table', {'class':'equation'})
    for form in formula:
        rs = form.findAll('span', class_='style5')
        for r in rs:
            y.append(r.text)
    #print(y)
    #print(y[0], end='\n')
    print()
```
